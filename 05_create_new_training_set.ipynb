```python
# Delete categorical variables from training dataset
training.drop(['ethnicity', 'gender', 'hospital_admit_source', 'icu_admit_source', 'icu_stay_type', 'icu_type'], axis=1, inplace=True)
# del training['ethnicity']
# del training['gender']
# del training['hospital_admit_source']
# del training['icu_admit_source']
# del training['icu_stay_type']
# del training['icu_type']
```

```python
# Create a list of numerical vars to drop
numerical_vars_to_drop = [col for col in numerical_vars.columns]
numerical_vars_to_drop.pop(0)
len(numerical_vars_to_drop)
```

```python
training.head()
```

```python
# Delete numerical vars from training datset
training.drop(numerical_vars_to_drop, axis=1, inplace=True)
```

```python
# So we dropped 6 categorical variables and 83 numerical variables from the dataset
# So 180 - (6 + 83) = 91
# So we should be left with 91 columns
training.shape
```

```python
# Now let's check the training dataset to see if there are any features with missing values
# From our eda, there should only be 1 binary column with missing values
training.isnull().sum()
```

```python
[col for col in training.columns if training[col].isnull().any()]
```

```python
len(training.columns[training.isnull().any()])
```

```python
more_numerical_vars_to_drop = [col for col in training.columns if training[col].isnull().any()]
# We'll remove the 4th element from the list (gcs_unable_apache) since this is the binary variable 
# with only a few missing values
more_numerical_vars_to_drop.pop(3) 
```

```python
len(more_numerical_vars_to_drop)
```

```python
# Delete more numerical vars from training datset
training.drop(more_numerical_vars_to_drop, axis=1, inplace=True)
```

```python
# We should get 84 + 44 - 1 = 127 columns
merged_df = pd.merge(categorical_vars, numerical_vars, on="Unnamed: 0")
```

```python
# Now let's jojin this merged_df with training dataset
# We should get 127 + 19 -1 = 145 columns
new_training_df = pd.merge(training, merged_df, on = "Unnamed: 0")
```
