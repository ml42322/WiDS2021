```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
train = pd.read_csv("./TrainingWiDS2021/TrainingWiDS2021.csv")
```

```python
cat_vars = train.select_dtypes(include=['object'])
```

```python
cat_vars = cat_vars.fillna('Unknown')
```

```python
cat_vars
```

```python
enc = OneHotEncoder(handle_unknown='ignore',drop='first')
enc.fit(cat_vars)
```

```python
ohe_df = enc.transform(cat_vars).toarray()
```

```python
ohe_df = pd.DataFrame(ohe_df, columns=enc.get_feature_names(cat_vars.columns))
```
